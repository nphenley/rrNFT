```python

## TRAIT GENERATION

TOTAL_RAGERS = 1000   #change this to however many you want

traits = []

def createCombo():
    
    trait = {}
    trait["Face"] = random.choices(faces, faceweights)[0]
    trait["Skin"] = random.choices(skins, skinweights)[0]
    trait["Leg"] = random.choices(legs, legweights)[0]
    trait["Gloves"] = random.choices(gloves, gloveweights)[0]
    trait["Eyes"] = random.choices(eyes, eyeweights)[0]
    trait["Hat"] = random.choices(hats, hatweights)[0]
    trait["Body"] = random.choices(bodies, bodyweights)[0]
    trait["Background"] = random.choices(backgrounds, backgroundweights)[0]
    
    if trait in traits:
        return createCombo()
    else:
        return trait
    
for i in range(TOTAL_RAGERS):
    
    newtraitcombo = createCombo()
    
    traits.append(newtraitcombo)
```
